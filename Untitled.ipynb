```python
# Import the main data analysis library
import pandas as pd
# Import visualization libraries (we'll use these later)
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visualization style for better aesthetics
sns.set_style("whitegrid")

# Load the dataset (it should be in the same folder as this notebook)
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Display the first 5 rows to ensure it loaded correctly
print("Data Loaded Successfully. First 5 Rows:")
print(df.head())

# Check the data types and look for missing values
print("\nDataFrame Information:")
df.info()
```

```python
# List the columns to drop based on initial inspection (they contain only one unique value)
columns_to_drop = ['EmployeeCount', 'StandardHours', 'Over18']

# Drop the columns permanently from the DataFrame
df = df.drop(columns=columns_to_drop, axis=1)

print(f"Dropped {len(columns_to_drop)} columns.")
print(f"New shape of the DataFrame: {df.shape}")
```

```python
# Convert 'Attrition' from categorical (Yes/No) to numerical (1/0)
df['Attrition_Numeric'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Display the overall attrition rate
attrition_rate = df['Attrition_Numeric'].mean() * 100

print(f"\nOverall Attrition Rate: {attrition_rate:.2f}%")
```

```python
# Group by OverTime and calculate the mean of the numeric attrition column
overtime_attrition = df.groupby('OverTime')['Attrition_Numeric'].mean().mul(100).sort_values(ascending=False)

# Print the result
print("\nAttrition Rate by OverTime:")
print(overtime_attrition)

# Visualize the result using a bar plot
plt.figure(figsize=(6, 4))
sns.barplot(x=overtime_attrition.index, y=overtime_attrition.values, palette="viridis")
plt.title('Attrition Rate (%) by OverTime Status')
plt.ylabel('Attrition Rate (%)')
plt.xlabel('OverTime')
plt.show()
```

```python
# Group by Department and calculate the mean attrition rate
dept_attrition = df.groupby('Department')['Attrition_Numeric'].mean().mul(100).sort_values(ascending=False)

print("\nAttrition Rate by Department:")
print(dept_attrition)

# Visualize the result
plt.figure(figsize=(8, 5))
sns.barplot(x=dept_attrition.index, y=dept_attrition.values, palette="rocket")
plt.title('Attrition Rate (%) by Department')
plt.ylabel('Attrition Rate (%)')
plt.xlabel('Department')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create the figure and axes for the plot
plt.figure(figsize=(10, 6))

# Plot the MonthlyIncome distribution for employees who stayed (Attrition = 'No')
sns.kdeplot(df.loc[df['Attrition'] == 'No', 'MonthlyIncome'], 
            label='Stayed (No Attrition)', fill=True, alpha=0.6)

# Plot the MonthlyIncome distribution for employees who left (Attrition = 'Yes')
sns.kdeplot(df.loc[df['Attrition'] == 'Yes', 'MonthlyIncome'], 
            label='Left (Attrition)', fill=True, alpha=0.6)

plt.title('Monthly Income Distribution by Attrition Status')
plt.xlabel('Monthly Income ($)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Create a histogram to show how many people leave based on years of service
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', 
             multiple='stack', bins=20, palette={'Yes': 'red', 'No': 'green'})
plt.title('Attrition vs. Years at Company')
plt.xlabel('Years at Company')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Export the cleaned DataFrame to a new CSV file
# Index=False prevents writing the DataFrame row numbers to the CSV
df.to_csv('cleaned_hr_attrition_data.csv', index=False)

print("Cleaned data exported successfully to 'cleaned_hr_attrition_data.csv'")
```
